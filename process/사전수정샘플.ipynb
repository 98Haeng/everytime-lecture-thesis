```python
df = pd.read_csv('everytime_review_data2.csv')
df.head()
```

```python
df['리뷰_형태소_list'][0]
```

```python
reviews = df[['리뷰','리뷰_형태소_list']]
reviews.columns = ['리뷰','형태소']
reviews.head()
```

```python
reviews['형태소'] = [re.sub(r"[^ㄱ-ㅣ가-힣a-zA-Z\s0-9+-,]", '', s) for s in reviews['형태소']]
```

```python
reviews['new바이그램'] = 0
for i in range(len(reviews)):
    bigram_list = []
    a = reviews['형태소'][i]
    a = a.split(", ")
    for j in range(len(a)-1):
        bigram_list.append(a[j] +"" + a[j+1])
    s = ", ".join(bigram_list)
    reviews['new바이그램'][i] = s
```

```python
reviews['new트라이그램'] = 0
for i in range(len(reviews)):
    bigram_list = []
    a = reviews['형태소'][i]
    a = a.split(", ")
    for j in range(len(a)-2):
        bigram_list.append(a[j] +"" + a[j+1] + "" + a[j+2])
    s = ", ".join(bigram_list)
    reviews['new트라이그램'][i] = s
```

```python
def change(data):
    # 단어 하나씩
    for i in range(len(data)):
        # 현재 구조
        string = data['바꿀단어'][i]
        # 바이그램, 트라이그램에 따른 패턴 설정
        if string.count(',') ==1:
            pat = "({})\s({})".format(string.split(' ')[0], string.split(' ')[1])
        else:
            pat = "({})\s({})\s({})".format(string.split(' ')[0], string.split(' ')[1], string.split(' ')[2])
        # 완벽하게 짜지는 않았는데 대충 이렇게?
        new_string = re.sub(pat, data['패턴'][i])
        
    # return만 어떻게 하고 적용하면 될듯
    return new_string # 예시
```
